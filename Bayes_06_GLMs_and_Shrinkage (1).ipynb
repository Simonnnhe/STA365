### Q1: derive the following

1. $p(\boldsymbol \beta | \textbf{y}, \textbf{X}, \Sigma=\sigma^2 I)$ for the **linear regression model** likelihood proportional to $\exp \left(-{\frac {1}{2}}(\mathbf {y} -\mathbf{X\boldsymbol \beta})^\top \Sigma^{-1}(\mathbf {y} - \mathbf{X\boldsymbol\beta})\right)$ and a $\mathcal{MVN}(\boldsymbol\beta_0, \boldsymbol\Sigma_\beta)$ for $\boldsymbol\beta$

Answer: 

From the question we know that the likelihood
$$
p(\textbf{y} | \boldsymbol \beta, \textbf{X}, \Sigma=\sigma^2 I) \propto \exp ( -\frac{1}{2}(\mathbf{y - X \boldsymbol\beta})^T \Sigma^{-1}(\mathbf{y - X \boldsymbol\beta})
$$

Also the prior $\boldsymbol \beta \sim \mathcal{MVN}(\boldsymbol\beta_0, \boldsymbol\Sigma_\beta)$, from the lecture note this means: 
$$
p(\boldsymbol \beta_{p\times 1})= \frac{1}{(2\pi)^{p/2} |\boldsymbol\Sigma_\beta|^{1/2}} \exp \left(-\frac{1}{2}(\boldsymbol\beta - \boldsymbol\beta_0)^\top \boldsymbol\Sigma_\beta^{-1} (\boldsymbol\beta - \boldsymbol\beta_0)\right)
$$

$$
\propto \exp  \left(-\frac{1}{2}(\boldsymbol\beta - \boldsymbol\beta_0)^\top \boldsymbol\Sigma_\beta^{-1} (\boldsymbol\beta - \boldsymbol\beta_0)\right)
$$


By the Bayessian Theorem we have:
$$
p(\boldsymbol\beta \mid \mathbf{y}, \mathbf{X}, \Sigma) \propto p(\mathbf{y} \mid \mathbf{X}, \boldsymbol\beta) \cdot p(\boldsymbol\beta).
$$

Subtituting the value:
$$
p(\boldsymbol\beta \mid \mathbf{y}, \mathbf{X}) \propto \exp\left( -\frac{1}{2} (\mathbf{y} - \mathbf{X}\boldsymbol\beta)^\top \Sigma^{-1}(\mathbf{y} - \mathbf{X}\boldsymbol\beta) \right) \cdot \exp\left( -\frac{1}{2} (\boldsymbol\beta - \boldsymbol\beta_0)^\top \boldsymbol\Sigma_\beta^{-1} (\boldsymbol\beta - \boldsymbol\beta_0) \right)
$$

$$
\propto{} \exp \left(-{\frac {1}{2}}({\boldsymbol \beta^\top}\mathbf{X^\top}\Sigma^{-1} \mathbf{X\boldsymbol \beta} - 2{\boldsymbol \beta^\top}\mathbf{X^\top}\Sigma^{-1} \mathbf{y}) \right) \times \exp \left(-{\frac {1}{2}}(\boldsymbol\beta^\top \Sigma_\beta^{-1}\boldsymbol\beta - 2 \boldsymbol \beta^\top \Sigma_\beta^{-1} \boldsymbol\beta_0)\right)
$$

$$
= \exp \left(-{\frac {1}{2}}({\boldsymbol \beta^\top}\left(\mathbf{X^\top}\Sigma^{-1} \mathbf{X} + \Sigma_\beta^{-1} \right)\boldsymbol \beta - 2{\boldsymbol \beta^\top}\left(\mathbf{X^\top}\Sigma^{-1} \mathbf{y} + \Sigma_\beta^{-1} \boldsymbol\beta_0\right))\right)
$$

Then this is just the kernel of Multivariate Normal distribution whose density function is:
$$
p(\boldsymbol{x}) = \frac{1}{(2\pi)^{d/2} |\boldsymbol\Sigma|^{1/2}} 
\exp\left( -\frac{1}{2}(\boldsymbol{x} - \boldsymbol\mu)^\top \boldsymbol\Sigma^{-1} (\boldsymbol{x} - \boldsymbol\mu) \right)
$$

Consider expand the term inside the exp, we have
$$
\exp\left( -\frac{1}{2}(\boldsymbol{x} - \boldsymbol\mu)^\top \boldsymbol\Sigma^{-1} (\boldsymbol{x} - \boldsymbol\mu) \right) 
= \exp(-\frac{1}{2}(x^T \Sigma^{-1}x - x^T\Sigma^{-1}x \mu - \mu^T \Sigma^{-1}x + \mu^T\Sigma^{-1}\mu)) \\
= \exp(-\frac{1}{2}(x^T\Sigma^{-1} x-2x^T\Sigma^{-1} \mu + \mu^T \Sigma^{-1} \mu))
$$
The last term is just a constant so we may ignore that. So we just focus on 
$$
\exp(-\frac{1}{2}(x^T\Sigma^{-1} x-2x^T\Sigma^{-1} \mu))
$$
By just replacing x to $\boldsymbol \beta$, this looks very similar to the posterior we get above.

Compare two expression we get the parameter where 
$$
\Sigma^{-1} = X^T\Sigma^{-1}X + \Sigma_\beta^{-1}
$$
and
$$
\Sigma^{-1}\mu = X^T\Sigma^{-1}y + \Sigma_\beta^{-1} \beta_0
$$

Rearrange the we get the updated parameters.

So $$p(\boldsymbol \beta | \textbf{y}, \textbf{X}, \Sigma=\sigma^2 I) \sim \mathcal{MVN}(\mu,\Sigma)$$ where
$$
\Sigma = (X^T\Sigma^{-1}X + \Sigma_\beta^{-1})^{-1} = COV(\beta)
$$
$$
\mu = \Sigma(X^T\Sigma^{-1}y + \Sigma_\beta^{-1} \beta_0)
$$



2. The analytical posterior distribution for $\sigma^2$ the **error variance** of a **linear regression model** with **design matrix** $\mathbf{X}$ assuming $\sigma^2$ has an **inverse-gamma** prior distribution with parameters $\alpha^*$ and $\beta^*$ (unrelated to $\boldsymbol \beta$).

Answer: 

Since prior of $p(\sigma^2) \sim$ inverse gamma $(\alpha^\ast, \beta^\ast)$
$$
p(\sigma^2) \propto (\sigma^2)^{-\alpha^\ast -1} \exp (-\frac{\beta^\ast}{\sigma^2})
$$

Also we have likelihood $p(y|X,\beta,\sigma^2) \sim \mathcal{MVN} (X\beta, \sigma^2 I)$ 

so 
$$
p(\textbf{y}_{n\times 1}| \textbf{X}_{n\times p}, \boldsymbol \beta, \sigma^2) = (2\pi )^{-n/2}(\sigma^2)^{-1/2}\,\exp \left(-{\frac {1}{2}}(\mathbf {y} -\mathbf{X\boldsymbol \beta})^\top (\sigma^2)^{-1}(\mathbf {y} - \mathbf{X\boldsymbol\beta})\right)
$$

$$
\propto \left(\sigma^2\right)^{-n / 2} \exp \left(-\frac{1}{2 \sigma^2}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})^{\top}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})\right)
$$

By the Bayes' Theorem, 
$$
p(\sigma^2 | X,Y,\beta) \propto p(y_{n\times 1}|X,\beta,\sigma^2) \times p(\sigma^2)
$$

$$
\propto\left(\sigma^2\right)^{-n / 2} \exp \left(-\frac{1}{2 \sigma^2}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})^{\top}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})\right) \times\left(\sigma^2\right)^{-\alpha^*-1} \exp \left(-\frac{\beta^*}{\sigma^2}\right)
$$

combining term we have:
$$
(\sigma^2)^{-\frac{n}{2}} \times (\sigma^2)^{-\alpha^\ast - 1} = (\sigma^2)^{\frac{-n-2\alpha^\ast-2}{2}} = (\sigma^2)^{-(\frac{n}{2}+\alpha^\ast+1)}
$$

For the exponent part:
$$
\exp \left(-\frac{\frac{1}{2}}{\sigma^2}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})^{\top}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})\right) \times \exp\left(-\frac{\beta^*}{\sigma^2}\right) = \exp(-\frac{\frac{1}{2} (y- X \beta)^T (y-X\beta) + \beta^\ast}{\sigma^2})
$$

So 
$$
p(\sigma^2 | X, Y ,\beta ) \propto (\sigma^2)^{-(\frac{n}{2}+\alpha^\ast+1)} \exp(-\frac{\frac{1}{2} (y- X \beta)^T (y-X\beta) + \beta^\ast}{\sigma^2})
$$

This is just the kernel of inverse Gamma distribution with updated parameter

$$
\alpha_{new} = \alpha^*+\frac{n}{2}
$$

$$
\beta_{new} = \frac{1}{2} (\mathbf{y}-\mathbf{X} \boldsymbol{\beta})^T (\mathbf{y}-\mathbf{X} \boldsymbol{\beta}) + \beta^{\ast}
$$

The data set I choose is sourced from Kaggle and it is about the BMI analysis(which might be seen very often among data analysis field). The data contains the height, gender, weight and the BMI index of about 741 individuals.  

```python
import pandas as pd
import pymc as pm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import arviz as az
from sklearn.preprocessing import StandardScaler
```

```python
#data sourse: 
#https://www.kaggle.com/datasets/rukenmissonnier/age-weight-height-bmi-analysis/data

df = pd.read_csv("bmi.csv")
df['Height'] = df['Height'] * 100
```

```python
df.shape
```

```python
# Define predictor variables and target variable
X = df[['Age', 'Height', 'Weight']].values
y = df['Bmi'].values
```

In this data set we would use age, height and weight to predict the BMI index.

```python
# Standardize predictors and add intercept
X_transformed = X.copy()
X_transformed[:, 0] = np.log1p(X[:, 0])  # Log(Age), use log1p to avoid log(0)
X_transformed[:, 2] = np.log1p(X[:, 2])  # Log(Weight)

scaler = StandardScaler()
X_scaled1 = scaler.fit_transform(X_transformed)
#X_new = np.hstack([np.ones((X.shape[0], 1)), X_scaled])#creates a column of ones, one for each observation.
# Add intercept to the designed matrix
#X_with_intercept = np.hstack([np.ones((X_1.shape[0], 1)), X_1])  # Intercept + predictors
n, p = X_scaled1.shape

#scaler_y = StandardScaler()
#y_scaled = scaler_y.fit_transform(y.reshape(-1, 1)).flatten()
#This just stores the number of rows n (observations) 
#and columns p (features + intercept) for later use in model defn.
#y_scaled = StandardScaler().fit_transform(y.reshape(-1, 1)).flatten()
```

```python
# Bayesian Linear Regression model
np.random.seed(1007691348)
with pm.Model() as model:
    beta0 = pm.Normal('beta0',mu=0,sigma=10)
    sigma = pm.HalfNormal('sigma', sigma=10)  # Non-inverse-gamma prior
    #betas = pm.StudentT('betas', mu=0, sigma=10,nu=3 shape=p)  # Scalar normal priors for coefficients
    betas = pm.Normal('betas', mu=0, sigma=100, shape=p)
    mu =    beta0+pm.math.dot(X_scaled1, betas)  # Linear predictor
    y_obs = pm.Normal('y', mu=mu, sigma=sigma, observed=y)  # Likelihood
    #y_obs = pm.StudentT('y', mu=mu, sigma=sigma, nu=3, observed=y)
 # Sampling
    idata1 = pm.sample(5000, tune=1000, return_inferencedata=True)
    
    
    
    
    
```

```python
# Summary of the posterior
summary = az.summary(idata1)
print(summary)

# Trace plots
az.plot_trace(idata1,combined=True)
plt.tight_layout()

#energy
az.plot_energy(idata1)
    
```

The trace plots and posterior distributions confirm that the Bayesian linear regression model is well-behaved. All parameters exhibit well-mixed MCMC chains with no divergences and with stable sampling. The posterior distributions are smooth and unimodal, with beta0 (intercept) centered around 26.4. However the trace plots for the regression coefficients appear visually flat, showing very small uncertainty. The residual standard deviation sigma is well-estimated around 1.52, indicating good model fit. Overall, the diagnostics support the reliability of the posterior estimates. Although we replaced the normal prior with a StudentT prior on the regression coefficients, the posterior means and trace plots remained nearly unchanged. This indicates that the data is highly informative, and the likelihood dominates the influence of the prior. The energy diagnostic plot shows good overlap between marginal energy and energy transition distributions, indicating smooth and efficient exploration of the posterior distribution. The BFMI values for all chains exceed 1.0, well above the acceptable threshold (0.3), confirming that the NUTS sampler performed efficiently. These results further suggest no concerns regarding divergent transitions or sampling inefficiencies.

The Bayesian linear regression model showed that age, height, and weight were all significant predictors of BMI, with weight having the strongest positive influence and height having a strong negative association. The intercept (mean BMI when predictors are zero) was around 26.36 (but this generally is meaningless as it is not real to have age, height, and weight all 0). The residual standard deviation (σ ≈ 1.52) indicated low unexplained variation in the outcome.

```python
p_3 = X_3.shape[1]
n_3 = X_3.shape[0]
```

```python
with pm.Model() as robust_model:
    # Priors
    betas = pm.Normal('betas', mu=0, sigma=10, shape=p_3)
    tau = pm.Gamma('tau', alpha=1.0, beta=1.0)
    nu = pm.Exponential('nu', lam=1/10)

    # Observation-specific precisions
    tau_i = pm.Gamma('tau_i', alpha=nu/2, beta=nu/2, shape=n_3)
    sigma_i = pm.Deterministic('sigma_i', 1 / pm.math.sqrt(tau * tau_i))

    # Linear predictor
    mu = pm.math.dot(X_3, betas)

    # Likelihood
    y_obs3 = pm.Normal('y', mu=mu, sigma=sigma_i, observed=y_outlier)

    # Sample from posterior
    trace = pm.sample(2000, tune=1000, target_accept=0.85, return_inferencedata=True)
```

```python
summary = az.summary(trace, var_names=['betas', 'tau', 'nu'], round_to=3)
print(summary)

#Trace plots
az.plot_trace(trace, var_names=['betas', 'tau', 'nu'])
plt.tight_layout()
plt.show()

# STEP 6: Energy diagnostic
az.plot_energy(trace)
plt.tight_layout()
plt.show()
```

The robust Bayesian linear regression model successfully accounts for the presence of outliers in the BMI dataset. The degrees of freedom (nu ≈ 0.65) indicates that the model prefers a heavy-tailed residual distribution. The global precision (tau ≈ 60) reflects moderate residual variance overall. The trace plots show good convergence across chains. One of the regression coefficients (beta2) appears disproportionately large, likely due to predictor scaling. However, model sampling behavior remains stable and reliable. The robust formulation allows automatic downweighting of outliers, resulting in more accurate and credible posterior estimates.
To evaluate sampling efficiency, energy diagnostics were performed using the Bayesian Fraction of Missing Information (BFMI). The BFMI values across the four chains ranged from 0.74 to 0.82. While all values were above 0.7, they fall slightly below the ideal threshold of 0.9, which indicates that there is moderate but acceptable sampling efficiency.  The marginal energy and energy transition distributions were reasonably aligned, suggesting that the sampler was able to explore the posterior distribution adequately. Despite this, no divergent transitions were observed, and all $\hat{R}$ values were approximately 1.0 with sufficient effective sample sizes, confirming that the posterior estimates remain reliable.
